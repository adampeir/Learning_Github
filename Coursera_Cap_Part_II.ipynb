Specifying URL

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
page = urllib.request.urlopen(url)
```

Examining HTML

```python
soup = BeautifulSoup(page, "lxml")
print(soup.prettify())
```

```python
soup.title.string

```

```python
all_tables=soup.find_all("table")
all_tables
```

```python
right_table=soup.find('table', class_='wikitable')
right_table
```

Scrapping data from the the three columns

```python
A=[]
B=[]
C=[]


for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
      
```

Cleaning data

```python

df=pd.DataFrame(A,columns=['PostalCode'])
df['Borough']=B
df['Neighborhood']=C
df = df.replace(r'\n',' ', regex=True) 
df = df[~df.Borough.str.contains("Not assigned")]
df = df.reset_index(drop=True)
df['Neighborhood'].replace(' /',', ',inplace=True, regex=True)
df['PostalCode'] = df['PostalCode'].astype(str)

```

```python
def test(x):
    if x == 'Not assigned':
        return True
    else:
        return False

for idx, val in enumerate(df.Neighborhood):
    if test(x) == True:
        df.Neighborhood[x] = df.Borough[x]
        
df
```

import Kmeans

```python
from sklearn.cluster import KMeans
```

Kmeans with 5 clusters

```python
kclusters = 5

X = results[['Latitude', 'Longitude']]

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(X)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# add clustering labels
results.insert(0, 'Cluster Labels', kmeans.labels_)


results.head() # check the last columns!
```

Create map with clusters

```python
import folium
import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as colors
# create map
map_clusters = folium.Map(location=[43.6532, -79.3832], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(results['Latitude'], results['Longitude'], results['Neighborhood'], results['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
